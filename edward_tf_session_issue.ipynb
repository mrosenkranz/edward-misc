```python
import edward as ed
import tensorflow as tf
from edward.models import Normal
```

```python
tf.reset_default_graph()
sess = tf.InteractiveSession()
```

```python
x = Normal(loc=[2.0], scale=[1.0], name='x')
qx = Normal(loc=tf.get_variable('qx/loc', [1]), scale=tf.nn.softplus(tf.get_variable('qx/scale', [1])))
```

```python
inference = ed.KLqp({x: qx})
inference.run(logdir=r'C:\tmp\log')
```
